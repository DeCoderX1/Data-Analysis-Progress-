```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "university": [
        "Anna University", "SRM University", "MIT", None, "VIT",
        "Anna University", "SRM Univ", "IIT Delhi", "VIT", "Private Univ"
    ],
    "type": [
        "Government", "Private", "Government", "Private", None,
        "Government", "Private", "Government", "Private", "Private"
    ],
    "applications": [
        12000, "15000", 14000, None, 18000,
        12000, "15500", 16000, None, "9000"
    ],
    "admissions": [
        6000, "7000", None, 3000, "8000",
        6000, 7100, None, 8200, "4500"
    ],
    "ranking": [
        12, "25", None, 40, 15,
        12, "26", 8, None, 55
    ]
}
data
```

```python
import pandas as pd
df = pd.DataFrame( {
    "university": [
        "Anna University", "SRM University", "MIT", None, "VIT",
        "Anna University", "SRM Univ", "IIT Delhi", "VIT", "Private Univ"
    ],
    "type": [
        "Government", "Private", "Government", "Private", None,
        "Government", "Private", "Government", "Private", "Private"
    ],
    "applications": [
        12000, "15000", 14000, None, 18000,
        12000, "15500", 16000, None, "9000"
    ],
    "admissions": [
        6000, "7000", None, 3000, "8000",
        6000, 7100, None, 8200, "4500"
    ],
    "ranking": [
        12, "25", None, 40, 15,
        12, "26", 8, None, 55
    ]
})
df
```

```python
# fix text coulmns
df['university']= df["university"].fillna("Unknown").str.strip().str.title()
df
```

```python
df["type"] = df['type'].fillna("Private").str.strip().str.title()
df
```

```python
# convert numeric columns
numeric_cols = ["applications",'admissions','ranking']
for col in numeric_cols:
    df[col] =pd.to_numeric(df[col], errors='coerce')
col
```

```python
df.head()   # to change ( integer == NaN , 12.0 == NaN ) like this asign !
```

```python
# remove duplicates
df.drop_duplicates()
df
```

```python
df.isnull()       # True represent ( Null Value )
```

```python
# fix missing numeric values
for col in numeric_cols:
    df[col]=df[col].fillna(df[col].median())   # you did see Nan - replace asign ( MEDIAN value)
df.head(6)
```

```python
# remove incorrect value ( greater than 100 )
df['ranking'] =df['ranking'].apply( lambda x: 100 if x>100 else x)
df.head(5)
```

```python
# application rate - new coulmns
df['ad_rate']=(df['admissions'] /df['applications'] * 100)
df.head(5)
```

```python
# replace Correct University name  ( you can desire name is incorrect - always change correct)
df['university'] = df['university'].replace({"Srm Univ" : "Srm University"})
df.head(5)
```

```python
df['university'] = df['university'].replace({"Private Univ" : "Private University"})
df
```

```python
# 8. Create category  high demand university
df["high_demand"] = df["applications"] > df["applications"].median()
df
```

```python
df["high_demand"] = df["high_demand"].map({True: "High Demand", False: "Low Demand"})
df.head(5)
```

```python
# to change ( gvt - pvt )
df["type"] = df["type"].replace({
    "Govt": "Government",
    "Pvt": "Private"
})
df.head(5)
```

```python
# sort dashboards View 
df = df.sort_values("applications", ascending=False)
df
```

# Visulaization :

```python
plt.figure(figsize=(8,4))
plt.bar(df["university"], df["applications"])
plt.xticks(rotation=45)
plt.title("University Applications Count")
plt.tight_layout()
plt.show()
```

```python
#Chart 2: Pie Chart  Share of Govt vs Private Universities
counts = df["type"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%")
plt.title("Government vs Private University Share")
plt.show()
```

```python
#Chart 3: Line Chart  Admissions Trend
plt.figure(figsize=(8,4))
plt.plot(df["university"], df["admissions"], marker='o')
plt.xticks(rotation=45)
plt.title("Admissions Trend Across Universities")
plt.tight_layout()
plt.show()
```

```python
# Chart 4: Scatter  Applications vs Admissions
plt.figure(figsize=(6,4))
plt.scatter(df["applications"], df["admissions"])
plt.xlabel("Applications")
plt.ylabel("Admissions")
plt.title("Applications vs Admissions")
plt.show()
```

```python
# Chart 5: Bar Chart  Admission Rate (%)
plt.figure(figsize=(8,4))
plt.bar(df["university"], df["admission_rate"])
plt.xticks(rotation=45)
plt.title("Admission Rate % by University")
plt.tight_layout()
plt.show()
```
